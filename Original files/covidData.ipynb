```python
import pandas as pd

data = 'owid-covid-data.csv'

covid_data = pd.read_csv(data)

covid_data.head()
```

```python
covid_data.shape
```

```python
pop_density_df = covid_data[['continent', 'location', 'date', 'total_cases', 'total_deaths', 'population_density']]

pop_density_df = pop_density_df.dropna()
pop_density_df.shape
```

```python
pop_density_df.groupby("location").count()
```

```python
hosp_bed_df = covid_data[['continent', 'location', 'date', 'total_cases', 'total_deaths', 'hospital_beds_per_thousand']]
hosp_bed_df = hosp_bed_df.dropna()
hosp_bed_df.shape
```

```python
hosp_bed_df.groupby("location").count()
```

```python
# How does the average life expect of a country correlate to total deaths
life_expectancy_df = covid_data[['continent', 'location', 'date', 'total_cases', 'total_deaths', 'life_expectancy']]
life_expectancy_df = life_expectancy_df.dropna()
life_expectancy_df.shape
```

```python
life_expectancy_df.groupby("location").count()
```

```python
# How does the total tests per thousand correlate to the total case numbers?
total_tests_df = covid_data[['continent', 'location', 'date', 'total_cases', 'total_tests_per_thousand']]
total_tests_df = total_tests_df.dropna()
total_tests_df.shape
```

```python
total_tests_df.groupby("location").count()
```

```python
# Total vaccination rate Vs new cases?
vaccination_df = covid_data[['continent', 'location', 'date', 'total_cases','new_cases', 'total_vaccinations', 'people_vaccinated', 
                                            'people_fully_vaccinated', 'total_boosters', 'new_vaccinations', 'total_vaccinations_per_hundred', 'people_vaccinated_per_hundred',
                                            'people_fully_vaccinated_per_hundred', 'total_boosters_per_hundred']]

vaccination_df.info()
```

```python
vaccination_df = vaccination_df.dropna()
vaccination_df.groupby("location").count()
```

```python
# New cases vs Government stringency 
stringency_df = covid_data[['continent', 'location', 'date', 'total_cases', 'new_cases', 'stringency_index']]
stringency_df.info()
```

```python
stringency_df = stringency_df.dropna()
stringency_df.groupby("location").count()
```
